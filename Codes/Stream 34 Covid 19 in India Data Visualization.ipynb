## Covid 19 in India Data Visualization using Plotly

```python
import pandas as pd
import plotly.graph_objs as go
from plotly.subplots import make_subplots
```

```python
data = pd.read_csv('../Covid_India.csv')
```

```python
data.head(5)
```

```python
fig = go.Figure()

fig.add_bar(x = data['Name of State / UT'], y = data['Total Confirmed cases'][:100])
```

```python
fig = go.Figure()

fig.add_trace(go.Pie(labels = data['Name of State / UT'][100:150],
                     values = data['Total Confirmed cases'][100:150]))
```
