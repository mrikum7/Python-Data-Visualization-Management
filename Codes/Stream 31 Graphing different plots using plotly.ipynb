```python
import plotly.graph_objs as go
import numpy as np
```

```python
fig = go.Figure()
x = [0.01 * i for i in range(-500, 501)]   ## List Comprehension
y = [np.sin(i) * np.cos(i) for i in x]

fig.add_trace(go.Scatter(x = x, y = y,
                         line = {'width' : 2,
                                 'color' : '#083d77'},
                         marker = {'size' : 6,
                                   'color' : "#da4167"},
                         mode = 'lines+markers'))
fig.update_xaxes(showline = True,
                 title = '<b>X Axis Data</b>',
                 title_font_family = 'Raleway',
                 title_font_color = 'Red',
                 mirror = True,
                 linewidth = 2,
                 zeroline = False,
                 showgrid = False,
                 linecolor = 'black')

fig.update_yaxes(showline = True,
                 mirror = True,
                 title = '<b>Y Axis Data</b>',
                 title_font_family = 'Raleway',
                 title_font_color = 'Red',
                 showgrid = False,
                 linewidth = 2,
                 zeroline = False,
                 linecolor = 'black')

fig.update_layout(plot_bgcolor = '#f4d35e')
```
