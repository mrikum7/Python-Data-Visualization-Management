```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import numpy as np
```

```python
fig = go.Figure()
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
z = np.cos(x)
fig.add_trace(go.Scatter3d(x = x, y = y, z = z,
                           marker = {'color' : '#c8553d',
                                     'size' : 3,
                                     'symbol' : 'circle'},
                           line = {'width' : 1, 
                                   'color' : '#588b8b'},
                          mode = 'markers+lines'))

```

```python
go.scatter3d.Line
```

```python
fig = make_subplots(rows = 2, cols = 2,
                    specs = [[{'type' : 'xy'}] * 2,
                             [{'type' : 'scene'}, {'type' : 'xy'}]])

fig.add_trace(go.Scatter(x = x, y = y), row = 1, col = 1)
fig.add_trace(go.Scatter(x = x, y = z), row = 1, col = 2)
fig.add_trace(go.Scatter3d(x = x, y = y, z = z), row = 2, col = 1)
fig.add_trace(go.Bar(x = [1, 2, 3, 4, 5], y = [2, 2, 4, 1, 5]), row = 2, col = 2)
```
